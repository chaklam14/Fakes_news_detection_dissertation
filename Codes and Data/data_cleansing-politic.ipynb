```python
import pandas as pd
dfr = pd.read_csv("politifact_real.txt")
dfr = dfr.reset_index(drop=True)

dff = pd.read_csv("politifact_fake.txt")
dff = dff.reset_index(drop=True)
```

```python
# fake = 1, real = 0
dfr['class'] = 0
dff['class'] = 1
```

```python
len(dfr.index)
#len(dff.index)
```

```python
# combine into one data set
df = pd.concat([dfr,dff])
df = df.reset_index()
```

```python
df.describe()
# df.tail(50)
# df.head(30)
# len(df.index)
```

```python
columns = df.columns
print(columns)
```

```python
import string
import re
import numpy as np

title = df[["title"]]
```

```python
# Finding numbers of special characters
sp = []
number = range(0, len(df.index))

for i in number:
    a = title.values[i]
    a = str(a)
    print(a)
    b = sum(bool(re.match(r"""[!><:;‘’“”'@#~{}\-_+=^&()?"]""", c)) for c in a)
    # each count need to subtract 2 because of the '' for each title
    b = b-2
    print(b)
    sp.append(b)
    
df['sp_count'] = sp
```

```python
# Finding numbers of CAPITAL

up = []
   
for i in number:
    a = title.values[i]
    a = str(a)
    print(a)
    count = sum(1 for c in a if c.isupper())
    a_up = count
    print(a_up)
    up.append(a_up)
    
df['up_count'] = up
```

```python
# Title length in CHARACTERS, not counting space

cha = []
   
for i in number:
    a = title.values[i]
    a = str(a)
    print(a)
    f = len(a)
    #f = f - 2 
    # print(f)
    f = f - 4 - a.count(' ') #minus '' and the space
    print(f)
    cha.append(f)

df['title_cha'] = cha
```

```python
# Title length in WORDS, not counting space

word = []
   
for i in number:
    a = title.values[i]
    a = str(a)
    a = a.split()
    print(a)
    b = len(a)
    print(b)
    word.append(b)

df['title_word'] = word
```

```python
# average number of characters of each word
a = []
cha_of_word = []
h1 = []
h2 = []
up_count = df[["up_count"]]
title_cha = df[["title_cha"]]
title_word = df[["title_word"]]
   
for i in number:
    h1 = title_cha.values[i]
    h2 = title_word.values[i]
    a = h1/h2
    a = a.round(decimals=1)
    a = float(a)
    cha_of_word.append(a)

df['cha_word'] = cha_of_word
```

```python
# Finding percentage of uppercase as character or word 

per = []
per2 = []

for i in number:
    g = up_count.values[i]
    hc = title_cha.values[i]
    hw = title_word.values[i]
    
    j = (g/hc)*100 # % in character
    k = (g/hw)*100 # % in word
    j = int(j)
    k = int(k)
    
    per.append(j) #character
    per2.append(k) #word

df['up_cha'] = per #character
df['up_word'] = per2 #word
```

```python
# BAD words
# dummy variable: 0 for no bad word, 1 for having bad words

bad = []

number = range(0, len(df.index))
for i in number:
    n = title.values[i]
    n = str(n)
    n = n.lower()
    n = n.split()
    print(n)
    if any(word in n for word in keyword):
        m = 1
    else:
        m = 0
    print(m)
    bad.append(m)
    
df['bad_word'] = bad
```
