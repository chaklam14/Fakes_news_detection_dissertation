```python
import pandas as pd
dfr = pd.read_csv("gossipcop_real.txt")
dfr = dfr.reset_index(drop=True)

dff = pd.read_csv("gossipcop_fake.txt")
dff = dff.reset_index(drop=True)
```

```python
# fake = 1, real = 0
dfr['class'] = 0
dff['class'] = 1
```

```python
# combine into one data set
df = pd.concat([dfr,dff])
df = df.reset_index()
```

```python
import string
import re
import numpy as np
title = df[["title"]]
```

```python
# Finding numbers of special characters
sp = []

number = range(0, len(df.index))
for i in number:
    a = title.values[i]
    a = str(a)
    print(a)
    b = sum(bool(re.match(r"""[!><:;‘’“”'@#~{}\-_+=^&()?"]""", c)) for c in a)
    # each count need to subtract 2 because of the '' for each title
    b = b-2
    print(b)
    sp.append(b)
    
df['sp_count'] = sp
```

```python
# Finding numbers of CAPITAL LETTER

up = []
   
for i in number:
    d = title.values[i]
    d = str(d)
    print(d)
    count = sum(1 for c in d if c.isupper())
    d_up = count
    print(d_up)
    up.append(d_up)
    
df['up_count'] = up
```

```python
# Title length CHARACTERS, not counting space

cha = []
   
for i in number:
    a = title.values[i]
    a = str(a)
    print(a)
    f = len(a)
    f = f - 4 - a.count(' ') #minus '' and the space
    print(f)
    cha.append(f)

df['title_cha'] = cha
```

```python
# Title length WORDS

word = []
   
for i in number:
    e = title.values[i]
    e = str(e)
    print(e)
    res = len(e.split()) 
    print(res)
    word.append(res)

df['title_word'] = word
```

```python
# average no. of characters of each word
a = []
cha_of_word = []
h1 = []
h2 = []
up_count = df[["up_count"]]
title_cha = df[["title_cha"]]
title_word = df[["title_word"]]
   
for i in number:
    h1 = title_cha.values[i]
    h2 = title_word.values[i]
    a = h1/h2
    a = a.round(decimals=1)
    a = float(a)
    cha_of_word.append(a)

df['cha_word'] = cha_of_word
```

```python
# Finding percentage of uppercase as character & words 

per = []
per2 = []

for i in number:
    g = up_count.values[i]
    hc = title_cha.values[i]
    hw = title_word.values[i]
    
    j = (g/hc)*100 #% in character
    k = (g/hw)*100 #% in word
    j = int(j)
    k = int(k)
    
    per.append(j)
    per2.append(k)

df['up_cha'] = per
df['up_word'] = per2
```

```python
# import bad words
dirty = pd.read_csv("DirtyWords.csv")
keyword = dirty['word']
dirty.describe()
```

```python
# BAD words
# dummy variable: 0 for no bad word, 1 for having bad words

bad = []

number = range(0, len(df.index))
for i in number:
    n = title.values[i]
    n = str(n)
    n = n.lower()
    n = n.split()
    print(n)
    if any(word in n for word in keyword):
        m = 1
    else:
        m = 0
    print(m)
    bad.append(m)
    
df['bad_word'] = bad
```
